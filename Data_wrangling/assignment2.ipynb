import the necessary libraries

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

import data

```python
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}
url = 'https://github.com/justmarkham/DAT8/blob/master/data/u.user'

r = requests.get(url)
soup = BeautifulSoup(r.content,'html5lib')
```

```python
print(soup.prettify())
```

```python
table = soup.find_all('td',attrs={'class':'blob-code blob-code-inner js-file-line'})
```

```python
table
```

```python
header = table[0].text
```

```python
columns = header.split('|')
```

```python
d = {}
```

```python
for column in columns:
    d[column] = []
```

```python
d
```

```python
for row in table[1:]:
    d['user_id'].append(int(row.text.split('|')[0]))
    d['age'].append(int(row.text.split('|')[1]))
    d['gender'].append(row.text.split('|')[2])
    d['occupation'].append(row.text.split('|')[3])
    d['zip_code'].append(row.text.split('|')[4])
```

```python
d
```

assign it to a variable called users and use the 'user_id' as index. 

```python
users = pd.DataFrame(data=d,index=d['user_id'])
```
