# 3. Low Level Data Understanding

***a. Find count of Unique values in 'location' column***


```python
len(df['location'].value_counts())
len(pd.unique(df['location']))
```

**b. Find which continent has maximum frequency using values count**

```python
df['continent'].value_counts()[[0]]
```

***c. Find maximum and mean value in 'Total Cases'***

```python
print('maximum no. of total cases:',df['total_cases'].max())
print('mean no. of total cases:',df['total_cases'].mean())
```

***d. Find 25%.50% & 75% quartile value in 'total deaths'***

```python
df['total_deaths'].quantile([0.25,0.50,0.75])
```

***e. Find which continent has maximum 'human development index'***


```python
df2 = df.groupby('continent')['human_development_index'].mean().reset_index()
df2.sort_values(by=['human_development_index'],ascending=False).reset_index().drop(columns=['index'])
```

***f. Find which continent has minimum 'GDP per Capita'***

```python
df3 = df.groupby('continent')['gdp_per_capita'].mean().reset_index()
df3.sort_values(by='gdp_per_capita',ascending=False).reset_index().drop(columns=['index'])
```

# 4. Filter the dataframe with only this columns
['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']
# and update the dataframe

```python
df = df.filter(['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
df
```

# 5. Data Cleaning

***a. Remove all duplicate observations***

```python
df.drop_duplicates(inplace=True)
df
```

***b. Find missing values in all columns***


```python
df.isnull().all()
```

```python
df.isnull()
```

***c. Remove all observations where continent column value is missing***

```python
df.dropna(subset=['continent'])
```

***d. Fill all missing values with 0***

```python
df.fillna(0,inplace=True)
```

```python
df.isnull()
```

# 6. Date time format

***a. Convert date column in datetime format using pandas .to_datetime***

```python
df['date'] = pd.to_datetime(df['date'])
```

***b. Create new column month after extracting month data from date column.***

```python
df['month'] = df['date'].dt.month
```

# 7. Data Aggregation


*   Find max value in all columns using groupby funciton on 'Continent'
use reset index after applying groupby
*   Store the result in a new dataframe named 'df_groupby'


```python
df_groupby = df.groupby('continent').max().reset_index()
df_groupby
```

```python
df_groupby.drop(0,inplace=True)
```

# 8. Feature Engineering

***a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'***

```python
df_groupby['total_deaths_to_total_cases'] = np.round(df_groupby['total_deaths']/df_groupby['total_cases'],3)
```

```python
df_groupby['total_deaths_to_total_cases']
```
